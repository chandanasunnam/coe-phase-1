# Mini Capstone: Predict Student Scores
This notebook demostrate how to build a Simple Linear Regression model to predict student scores based on hours studied

```python
#Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
#step2: Load Dataset
url = "https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"

df=pd.read_csv(url)
df.head()
```

```python
#step 3 : visualize data
plt.scatter(df["Hours"], df["Scores"])
plt.title("Hours vs Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.grid(True)
plt.show()
```

```python
#step 4: Prepare Data
x=df[["Hours"]]
y=df["Scores"]
```

```python
#step 5: Train/Test Split
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#step 6: Train Linear Regression Model
model=LinearRegression()
model.fit(x_train, y_train)
```

```python
#step 7: predict
y_pred = model.predict(x_test)
```

```python
#step 8:Evalute Model
print("R2 Score:", r2_score(y_test, y_pred))
print("Mean Absolute Error:",mean_absolute_error(y_test,y_pred))
```

```python
#step 9: Visualize Regression Line
plt.scatter(x,y,color='blue')
plt.plot(x,model.predict(x), color='red')
plt.xlabel("Regression Line - Student Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.show()
```

```python
#step 10: predict for custom input
hours = [[8.25]]
predicted_score = model.predict(hours)
print(f"Predicted score for studying {hours[0][0]} hours = {predicted_score[0]:.2f}")
```
